# 图像绘制

```python
import matplotlib.pyplot as plt

# 1. 创建画布
plt.figure(figsize=(20,8), dpi=100)

# 2. 绘制图像
x = [1, 2, 3]
y = [4, 5, 6]
plt.plot(x, y)

# 3. 显示图像
plt.show()
```

# 图像保存

```python
# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2. 图像绘制
x = [1, 2, 3, 4, 5, 6]
y = [3, 6, 3, 5, 3, 10]
plt.plot(x, y)

# 3. 图像保存，一定要在 show() 前面
plt.savefig("./data/test.png")

plt.show()
```
