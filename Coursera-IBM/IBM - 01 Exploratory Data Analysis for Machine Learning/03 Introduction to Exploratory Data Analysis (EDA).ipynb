### Sampling from DataFrames

```python
import pandas as pd
import numpy as np

filepath = 'data/iris.data'
data = pd.read_csv(filepath, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Sample 5 rows without replacement
sample = data.sample(n=5, replace=False)
print(sample)
print(sample.iloc[:,-3:]) # print all rows ([:]) with last 3 columns ([-3:0])
```

## EDA with Visualization

* Matplotlib
* Pandas (via Matplotlib)
* Seaborn
  * Statistically-focused plotting methods
  * Global preferences incorporated by Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Pandas DataFrame approach
plt.plot(data.sepal_length, data.sepal_width, ls='', marker='o')
```

```python
# First plot statement
plt.plot(data.sepal_length, data.sepal_width, ls='', marker='o', label='sepal')

# Second plot statement
plt.plot(data.petal_length, data.petal_width, ls='', marker='x', label='petal')
```

### Customizing Plots

```python
# matplotlib syntax
fig, ax = plt.subplots()
ax.barh(np.arange(10), data.sepal_width.iloc[:10])

# Set position of ticks and tick labels
ax.set_yticks(np.arange(0.4, 10.4, 1.0))
ax.set_yticklabels(np.arange(1, 11))
ax.set(xlabel='xlabel', ylabel='ylabel', title='Title')
```

### Grouping Data for EDA

#### Customizing Plots: by Group

```python
# Pandas DataFrame approach
data.groupby('species').mean().plot(color=['red', 'blue', 'black', 'green'], fontsize=16.0, figsize=(8, 8))
```

#### Pair Plots for Features

```python
import seaborn as sns
```

```python
# Seaborn plot, feature correlations
sns.pairplot(data, hue='species', height=3)
```

### Seaborn Example: Hexbin Plot

```python
# Seaborn hexbin plot
sns.jointplot(x=data['sepal_length'],y=data['sepal_width'], kind='hex')
```

### Seaborn Example: Facet Grid

```python
# Seaborn plot, Facet Grid
# First plot statement
from turtle import color

plot = sns.FacetGrid(data, col='species', margin_titles=True)
plot.map(plt.hist, 'sepal_width', color='green')
# Second plot statement
plot = sns.FacetGrid(data, col='species', margin_titles=True)
plot.map(plt.hist, 'sepal_length', color='blue')
```
