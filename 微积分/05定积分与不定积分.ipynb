# 目的
* 找到原函数的**表达式**
  * **不定**积分 $\int f'(t)dt=f(t)+c$
* 求出一个**数值**，即是原函数在一个区间的变化量
  * **定**积分

# 问题
参考*02微分思想.ipynb*中的 时间|位移 例子。在时间$t_0$到$t_1$中有多少位移？

不定积分中，不论常数c怎么变，特定t区间的位移$f(t)$是固定的，因为$[f(t_1)+c]-[f(t_0)+c]=f(t_1)-f(t_0)$

因此，此处要计算的位移，即定积分
* $\int_{t_1}^{t_0}f'(t)dt=f(t_1)-f(t_0)$

```python
import numpy as np
import matplotlib.pyplot as plt

c = 1
def func1(x):
    return 3*x**2

def func2(x):
    return x**3+c

plt.figure(figsize=(16, 6))
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
xmin = 0
xmax = 3
ymin = 0
ymax = 4
plt.subplot(1, 2, 1)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

x = np.arange(0, 2, 0.1)
y1 = func1(x)
y2 = func2(x)
plt.plot(x, y1)

ax= plt.gca()

# 箭头坐标
for side in ['bottom', 'right', 'top', 'left']:
    ax.spines[side].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
xhead_width = 0.12
xhead_length = 0.15
yhead_width = 0.08
yhead_length = 0.25
line_width = 1.
overhang = .3
ax.arrow(0, 0, xmax, 0., fc='k', ec='k', lw=line_width, head_width=xhead_width, head_length=xhead_length, overhang=overhang, length_includes_head=True, clip_on=False)
ax.arrow(0, 0, 0, ymax, fc='k', ec='k', lw=line_width, head_width=yhead_width, head_length=yhead_length, overhang=overhang, length_includes_head=True, clip_on=False)
plt.xticks([3], ['x'])
plt.yticks([])
plt.annotate(r"$f'(x)=3x^2$", xy=(0, 4), xycoords='data', xytext=(+15, -15), textcoords='offset points', fontsize=16)

xa0 = fxa0 = 0
for xa in [.2, .4, .6, .8, 1]:
    fxa=func1(xa)
    plt.scatter(xa, fxa, s=50, color='red')
    plt.plot([xa, xa], [fxa, 0], 'k--', lw=1.0)
    plt.plot([xa0, xa], [fxa, fxa], 'k--', lw=1.0)
    plt.plot([xa0, xa0], [fxa0, fxa], 'k--', lw=1.0)
    xa0 = xa
    fxa0 = fxa
    # plt.annotate(r'$(1, 1)$', xy=(x1, fx1), xycoords='data', xytext=(+5, -5), textcoords='offset points', fontsize=16)

plt.annotate(r"小方块面积$=f'(dx)\times dx$", xy=(1, 2), xycoords='data', xytext=(+15, +5), textcoords='offset points', fontsize=16)

plt.subplot(1, 2, 2)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.plot(x, y2)
ax= plt.gca()

# 箭头坐标
for side in ['bottom', 'right', 'top', 'left']:
    ax.spines[side].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
xhead_width = 0.12
xhead_length = 0.15
yhead_width = 0.08
yhead_length = 0.25
line_width = 1.
overhang = .3
ax.arrow(0, 0, xmax, 0., fc='k', ec='k', lw=line_width, head_width=xhead_width, head_length=xhead_length, overhang=overhang, length_includes_head=True, clip_on=False)
ax.arrow(0, 0, 0, ymax, fc='k', ec='k', lw=line_width, head_width=yhead_width, head_length=yhead_length, overhang=overhang, length_includes_head=True, clip_on=False)
plt.xticks([3], ['x'])
plt.yticks([c], ['c'])
plt.annotate(r"$f(x)=x^3+1$", xy=(0, 4), xycoords='data', xytext=(+15, -15), textcoords='offset points', fontsize=16)

for xa in [1]:
    fxa=func2(xa)
    fxa0=func2(0)
    
    plt.scatter(xa, fxa, s=50, color='red')
    plt.scatter(0, fxa0, s=50, color='red')
    plt.plot([xa, xa], [fxa, fxa0], 'k', lw=2.0)
    plt.plot([0, xa], [fxa0, fxa0], 'k--', lw=1.0)
    xa0 = xa
    fxa0 = fxa
    
plt.annotate(r"$\Delta f(x)=f(x_0)-c$", xy=(1, 1.5), xycoords='data', xytext=(+15, 0), textcoords='offset points', fontsize=16)
plt.show()
```

上图：

将倒数图微分，计算每个矩形的面积，其面积和为$f'(dx)\times dx$，这与原函数在这个区间值的变化一致，即$f(x_1)-f(x_0)$

也就是$\int_a^b\Delta f(x)\rightarrow\int_a^bf'(x)dx$

# 从定积分到不定积分
* $\int_0^xf'(t)dt=f(x)-f(0)$  
  这里左边不用$x$了，叫做傀儡自变量（dummy variable）
* $-f(0)$是一个常数
* 不定积分的**结果**，使定积分的**运算**变得简单
* 定积分的运算，使一些**复杂**的**几何**问题变得**简单**
  * 曲线长度，曲面面积，不规则形体体积，等
