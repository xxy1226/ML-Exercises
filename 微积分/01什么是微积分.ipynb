# 什么是微积分

* **Calculus** 拉丁语： 小石子 --> 计算
* 把**无穷**引入计算后，产生的一套计算方法
* 把有限的东西**分割**无穷多次后观察，把无穷小**累积**后观察的行为
* 既然无穷小，还有什么意义？

## 无穷也能计算？

* 无穷和无穷不一样  
  1和2之间有无穷多个数，1.1, 1.11, 1.111, ...  
  2和3之间也有无穷多个数， 2.1, 2.11, 2.111, ...  
  那么1和3之间的数，比1和2之间多？还是少？还是一样多？
* 既然不一样，不同的无穷小可不可以比较
* 既然无穷小是从有限值里分割出来的，那么无穷小也可以累积为**有限值**

## 微积分有什么用/为什么要学微积分

* 微积分是**现代数学**的基础，进而是**现代科学**的基础
* 研究**变化**的规律
* 尤其与**物理**联系紧密
* 起源：研究*天体运动**

## 微积分难么

* **本质**并不难
* 有些**概念**比较抽象：**极限**，**连续**
* 相关知识：**三角函数**，**双曲函数**，**幂函数**，**指数函数**，等
* **运算**规则
* 陌生的语言，希腊字母
* **用就不难**
