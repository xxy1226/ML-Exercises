# 切线
* 一条**直线**
* 经过**光滑**去线上的某一点
* **没有其他的**交点
* 叫做这条曲线，在**这个点商**的切线
* 特征： **斜率 ($tan\alpha$)** (见下图)
* **唯一的**

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
xmin = 0
xmax = 3
ymin = 0
ymax = 4
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

x = np.arange(0, 2, 0.2)
y1 = x**2
y2 = 2*x-1
plt.plot(x, y1)
plt.plot(x, y2)

ax= plt.gca()

# 箭头坐标
for side in ['bottom', 'right', 'top', 'left']:
    ax.spines[side].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
xhead_width = 0.25
xhead_length = 0.3
yhead_width = 0.25
yhead_length = 0.5
line_width = 1.
overhang = .3
ax.arrow(0, 0, xmax, 0., fc='k', ec='k', lw=line_width, head_width=xhead_width, head_length=xhead_length, overhang=overhang, length_includes_head=True, clip_on=False)
ax.arrow(0, 0, 0, ymax, fc='k', ec='k', lw=line_width, head_width=yhead_width, head_length=yhead_length, overhang=overhang, length_includes_head=True, clip_on=False)


x1 = 1
fx1 = x1**2
plt.scatter(x1, fx1, s=50, color='red')
# plt.plot([t1, t1], [S1, 0], 'k--', lw=2.0)
# plt.plot([0, t1], [S1, S1], 'k--', lw=2.0)
plt.annotate(r'$(1, 1)$', xy=(x1, fx1), xycoords='data', xytext=(+5, -5), textcoords='offset points', fontsize=16)
plt.annotate(r'$\alpha$', xy=(0.5, 0), xycoords='data', xytext=(+15, +5), textcoords='offset points', fontsize=16)


plt.show()
```

将上图微分，那一点的斜率就是$\frac{df(x)}{dx}$，也就是导数
