上图是匀速直线运动的 时间|位移 关系

计算速度： $V=\frac{S_2-S_1}{t_2-t_1}$

```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
# from pandas import Series, DataFrame

plt.figure(num='时间与距离', figsize=(8, 6))
t = np.arange(0, 5, 0.2)
S = t**2
plt.plot(t, S)
plt.xlim(0, 6)
plt.ylim(0, 25)
ax= plt.gca()

# 箭头坐标
for side in ['bottom', 'right', 'top', 'left']:
    ax.spines[side].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
xhead_width = 1.25
xhead_length = 0.3
yhead_width = 0.25
yhead_length = 1.6
line_width = 1.
overhang = .3
ax.arrow(0, 0, 6, 0., fc='k', ec='k', lw=line_width, head_width=xhead_width, head_length=xhead_length, overhang=overhang, length_includes_head=True, clip_on=False)
ax.arrow(0, 0, 0, 25., fc='k', ec='k', lw=line_width, head_width=yhead_width, head_length=yhead_length, overhang=overhang, length_includes_head=True, clip_on=False)


t1 = 3.4
S1 = t1**2
plt.plot([t1, t1], [S1, 0], 'k--', lw=2.0)
plt.plot([0, t1], [S1, S1], 'k--', lw=2.0)
plt.annotate(r'$\Delta t$', xy=(t1, 3), xycoords='data', xytext=(+6, -30), textcoords='offset points', fontsize=16)
t2 = 3.8
S2 = t2**2
plt.plot([t2, t2], [S2, 0], 'k--', lw=2.0)
plt.plot([0, t2], [S2, S2], 'k--', lw=2.0)
plt.annotate(r'$\Delta S$', xy=(0, S2), xycoords='data', xytext=(+10, -25), textcoords='offset points', fontsize=16)

plt.annotate(r'$\bar{V}=\frac{\Delta S}{\Delta t}$', xy=(4, 15), xycoords='data', xytext=(+10, -30), textcoords='offset points', fontsize=20)

plt.xticks([t1, t2, 6], ["t1", "t2", "t"])
plt.yticks([S1, S2, 25], ["S1", "S2", "S"])


plt.show()
```

上图是一个加速运动，要计算一个时间点的瞬时速度

将时间点看做是一个无限接近于0的时间段 $\delta t$，只要算出这个时间段中的平均速度，我们就认为这是这个时间点的瞬时速度。这个速度和实际速度的区别就是误差。

```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
# from pandas import Series, DataFrame

plt.figure(num='时间与距离', figsize=(8, 6))
t = np.arange(0, 5, 0.2)
S = t**3
plt.plot(t, S)
plt.xlim(0, 6)
plt.ylim(0, 125)
ax= plt.gca()

# 箭头坐标
for side in ['bottom', 'right', 'top', 'left']:
    ax.spines[side].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
xhead_width = 6.25
xhead_length = 0.3
yhead_width = 0.25
yhead_length = 8
line_width = 1.
overhang = .3
ax.arrow(0, 0, 6, 0., fc='k', ec='k', lw=line_width, head_width=xhead_width, head_length=xhead_length, overhang=overhang, length_includes_head=True, clip_on=False)
ax.arrow(0, 0, 0, 125, fc='k', ec='k', lw=line_width, head_width=yhead_width, head_length=yhead_length, overhang=overhang, length_includes_head=True, clip_on=False)


t1 = 3.4
S1 = t1**3
plt.plot([t1, t1], [S1, 0], 'k--', lw=2.0)
plt.plot([0, t1], [S1, S1], 'k--', lw=2.0)
plt.annotate(r'$\Delta t$', xy=(t1, 15), xycoords='data', xytext=(+6, -30), textcoords='offset points', fontsize=16)
t2 = 3.8
S2 = t2**3
plt.plot([t2, t2], [S2, 0], 'k--', lw=2.0)
plt.plot([0, t2], [S2, S2], 'k--', lw=2.0)
plt.annotate(r'$\Delta f$', xy=(0, S2), xycoords='data', xytext=(+10, -25), textcoords='offset points', fontsize=16)

plt.annotate(r'$f(t)=t^3$', xy=(4, 115), xycoords='data', xytext=(+10, -30), textcoords='offset points', fontsize=20)

plt.xticks([t1, t2, 6], ["$t_0$", "$t_0+\Delta t$", "$t$"])
plt.yticks([S1, S2, 125], ["$f(t_0)$", "$f(t_0+\Delta t)$", "$f(t)$"])


plt.show()
```

$\Delta f=f(t_0+\Delta t)-f(t_0)$  
&emsp;&emsp;$=(t_0+\Delta t)^3-(t_0)^3$  
&emsp;&emsp;$=3t_0^2\times\Delta t+3(\Delta t)^2\times t_0+(\Delta t)^3$

$v=\Delta f/\Delta t=3t_0^2+3\Delta t\times t_0+(\Delta t)^2$  
&emsp;$\approx3t_0^2$

* $\Delta x\rightarrow dx$;&emsp;$\Delta f(x)\rightarrow df(x)$
* $\frac{df(x)}{dx}$：导数，derivative
* 运算过程：微分，differentiation
* 导数也可以写作 $f'$
